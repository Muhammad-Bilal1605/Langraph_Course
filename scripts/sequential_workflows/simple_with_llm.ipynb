```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate
from langgraph.graph import StateGraph,END, START
from typing import TypedDict
```

```python
OLLAMA_MODEL = "qwen2.5-coder:14b"
OLLAMA_BASE_URL = "http://localhost:11434"
```

```python
model = ChatOllama(
    model=OLLAMA_MODEL,
    base_url=OLLAMA_BASE_URL,
    temperature=0,
)
```

```python
class ConvType(TypedDict):
    question:str
    answer:str
```

```python
def conversation(state: ConvType):
    question = state["question"]

    template = PromptTemplate(
        input_variables=["question"],
        template="You are a helpful AI that gives short but engaging GenZ-style answers. Use slang. Question: {question}"
    )

    prompt = template.format(question=question)
    response=model.invoke(prompt).content

    return {"answer": response}
```

```python
graph= StateGraph(ConvType)

graph.add_node('conversation',conversation)
graph.add_edge(START,"conversation")
graph.add_edge("conversation",END)
```

```python
workflow=graph.compile()
workflow.invoke({'question':"what can you tell me about blackholes"})
```
