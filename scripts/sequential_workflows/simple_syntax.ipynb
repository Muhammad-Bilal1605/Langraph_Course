```python
from langgraph.graph import StateGraph, START,END
from typing import TypedDict,NotRequired
```

```python

class BmiState(TypedDict):
    weight: float
    height: float
    bmi: NotRequired[float]
    category: NotRequired[str]
```

```python
def calculate_bmi(state: BmiState):
    bmi = state['weight'] / (state['height'] ** 2)
    return {"bmi": bmi}
```

```python
def classify_category(state:BmiState):
    if 18.5 < state['bmi'] < 24.5:
        category="good"
    else:
        category="bad"
    return {'category':category}
```

```python
graph= StateGraph(BmiState)
graph.add_node('calculate_bmi',calculate_bmi)
graph.add_node('classify_category',classify_category)
```

```python
graph.add_edge(START,'calculate_bmi')
graph.add_edge('calculate_bmi','classify_category')
graph.add_edge('classify_category',END)
```

```python
workflow= graph.compile()
```

```python
workflow.invoke({"weight":80,"height":1.8})
```
